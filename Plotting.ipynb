```python
import numpy as np
import math
import scipy.interpolate
from matplotlib import pyplot
import matplotlib
import seaborn as sns
```

```python
tfont = {'fontname':'Times New Roman'}
```

```python
alpha = 1.
k_B = 8.6173303e-5
hbar = 6.582119514e-16
m = 1.
k = 1.
w = np.sqrt(k/m)
T = np.linspace(0., 1000., num=1000)
x21 = (k_B*T)/k
x22 = (x21 - ((15*alpha/4)*((k_B**2*T**2)/(k**3))))/(1-((3*alpha/4)*((k_B*T)/(k**2))))
```

```python
%matplotlib inline

sns.set_style("ticks", {'font.family': 'Times New Roman','axes.axisbelow': False,'grid.linestyle': u'--', 'grid.color': '.5', 'legend.frameon': True})

sns.set_palette(sns.color_palette('GnBu', n_colors=2))

x_start, x_end = T.min(), 1.1*T.max()
y_start, y_end = x21.min(), 1.1*x21.max()

pyplot.figure(1)
sns.set_context("talk", font_scale=2.5)
pyplot.figure(figsize=(12, 10))
pyplot.grid(True)
pyplot.ylabel(r'$\left<x^2\right>$ ($eV$)')
pyplot.xlabel(r'T ($K$)')
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.plot(T, x21, linestyle='solid')
pyplot.plot(T, x22, linestyle='solid')
pyplot.legend([r'$\alpha = 0$', r'$\alpha = 1$'], loc='best');
pyplot.savefig('1a.pdf', dpi=300, bbox_inches='tight');
```
